```python
import pandas as pd
data=pd.read_csv('Latest Covid-19 India Status.csv')
print(data.head())
```

```python
print(data.columns)
print(data.info())
```

**SIMPLE VISUALIZATION**

```python
import matplotlib.pyplot as plt

# BAR PLOT OF TOP 5 STATES WITH CASES
top_states=data.nlargest(5,'Total Cases')
plt.bar(top_states['State/UTs'],top_states['Total Cases'])
plt.xlabel('States')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```
